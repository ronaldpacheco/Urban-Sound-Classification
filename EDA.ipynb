```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
import seaborn as sns
```

```python
df = pd.read_csv('wrangled_data.csv')
```

```python
df.head()
```

```python
df.drop(columns ='Unnamed: 0', inplace=True)
df.info(verbose=True)
```

```python
df.hist(figsize=(50,50))
plt.show()
```

```python
pd.set_option('display.max_rows', 500)
df.describe()
```

```python
df.corr()
```

Let's create heatmaps to visualize the correlation better, just to get an idea of how different our sounds are:

### melspectrograms

```python
melspectrogram = [f'melspectrogram_{n}' for n in range(128)]
plt.figure(figsize=(20,20))
sns.heatmap(df[melspectrogram].corr())
df[melspectrogram].hist(figsize=(30,30), bins=50)
plt.show()
```

- We can see some high correlation amongst the melspectrogram features, however, most of them are not correlated.
- There are outliers that make our distribution right-skewed. Let's see which sounds are causing these outliers:



```python
df[df.melspectrogram_0 >3].T
```

I was expecting to see low variety in classes for these outliers, but it looks to be all over the place. Outliers for melspectrogram_0 are not the same for melspectrogram_60 (for example).

### mfcc:

```python
mfcc = [f'mfcc_{n}' for n in range(20)]
plt.figure(figsize=(10,10))
sns.heatmap(df[mfcc].corr())
df[mfcc].hist(figsize=(10,15))
plt.show()
```

- This is a little more uniform, mfcc_0 and mfcc_1 seem to not be correlated with any other mfcc feature. 
- They are all normally distributed.

### chroma_stft:

```python
chroma_stft = [f'chroma_stft_{n}' for n in range(12)]
plt.figure(figsize=(10,10))
sns.heatmap(df[chroma_stft].corr())
df[chroma_stft].hist(figsize=(10,10))
plt.show()
```

- Here, the features seem to be highly correlated with the its previous feature, and this seems to carry over from the last chroma feature to the first one (chroma_stft_0 & chroma_stft_11).
- They all are fairly normally distributed with a slight left-skew.

### spectral_contrast

```python
spectral_contrast = [f'spectral_contrast_{n}' for n in range(12)]
plt.figure(figsize=(10,10))
sns.heatmap(df[spectral_contrast].corr())
df[spectral_contrast].hist(figsize=(10,10))
plt.show()
```

- Looks exactly the same as chroma_stft. So we might choose to use only one of these two set of features.
- They all are fairly normally distributed with a slight left-skew.

### Remainder features

```python
l = spectral_contrast + melspectrogram + chroma_stft + ['Class', 'audio_name'] + mfcc
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(df.drop(columns=l).corr())
plt.show()
```

Here we see that zcr is highly correlated with spectral_centroid, while spectral_centroid, spectral_rolloff and spectral_bandwith are highly correlated amongst themselves.
